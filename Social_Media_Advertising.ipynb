## 2 - Feature Engineering

```python
df = df.dropna(subset=['Date'])
df = df.sort_values('Date')
df = df.drop_duplicates()

display(df)
```

### 2.1 - Create KPI Features

```python
import numpy as np

df['Conversions'] = df['Clicks'] * df['Conversion_Rate']

df['CTR'] = np.where(df['Impressions']>0, df['Clicks']/df['Impressions'], 0.0)
df['CPM'] = np.where(df['Impressions']>0, (df['Acquisition_Cost']/df['Impressions']) * 1000, 0.0)
df['CPA'] = np.where(df['Conversions']>0, df['Acquisition_Cost']/df['Conversions'], np.nan)
df['Engagement_Rate'] = np.where(df['Impressions']>0, df['Engagement_Score']/df['Impressions'], 0.0)

df['Spend_per_Day'] = np.where(df['Duration_days']>0, df['Acquisition_Cost'] / df['Duration_days'], np.nan)
df['Estimated_Revenue'] = df['Acquisition_Cost'] * (1 + df['ROI'])

df['ROI_values'] = ((df['Estimated_Revenue'] - df['Acquisition_Cost']) / df['Acquisition_Cost']) * 100


display(df)
```

### 2.2 - Create aggregate data

For each company , we will aggregate the data on daily basis for time series modelling input 

```python
parent_key = ['Company']
sum_cols = ['Acquisition_Cost','Clicks','Impressions','Engagement_Score','Conversions','Duration_days','Estimated_Revenue']

def aggregate_daily(df, keys):
    gcols = keys + ['Date']
    tmp = df[gcols + sum_cols].copy()

    agg = tmp.groupby(gcols).agg(
        {**{c:'sum' for c in sum_cols}}    
    ).reset_index()

    agg['CTR'] = np.where(agg['Impressions']>0, agg['Clicks']/agg['Impressions'], 0.0)
    agg['CPM'] = np.where(agg['Impressions']>0, (agg['Acquisition_Cost']/agg['Impressions']) * 1000, 0.0)
    agg['CPA'] = np.where(agg['Conversions']>0, agg['Acquisition_Cost']/agg['Conversions'], np.nan)
    agg['Engagement_Rate'] = np.where(agg['Impressions']>0, agg['Engagement_Score']/agg['Impressions'], 0.0)

    agg['Spend_per_Day'] = np.where(agg['Duration_days']>0, agg['Acquisition_Cost'] / agg['Duration_days'], np.nan)

    # agg['Estimated_Revenue'] = agg['Acquisition_Cost'] * (1 + agg['ROI'])
    agg['ROI_values'] = ((agg['Estimated_Revenue'] - agg['Acquisition_Cost']) / agg['Acquisition_Cost']) * 100


    # calendar
    agg['dow'] = pd.to_datetime(agg['Date']).dt.dayofweek
    agg['month'] = pd.to_datetime(agg['Date']).dt.month
    agg['is_weekend'] = agg['dow'].isin([5,6]).astype(int)

    return agg.sort_values(keys + ['Date'])

daily_data = aggregate_daily(df, parent_key)

display(daily_data)
```

```python
def add_lags_rolls(frame, target='ROI_values', feat_cols=None, lags=(1,7), rolls=(7,), keys=parent_key):

    out = frame.copy()
    group = out.groupby(keys) if len(keys)>0 else [((), out)]

    frames = []
    for _, g in group:
        g = g.sort_values('Date').copy()

        for L in lags:
            g[f'{target}_lag{L}'] = g[target].shift(L)
        for W in rolls:
            g[f'{target}_rollmean{W}'] = g[target].rolling(W, min_periods=max(3, W//2)).mean()

        for c in feat_cols:
            for L in lags:
                g[f'{c}_lag{L}'] = g[c].shift(L)
            for W in rolls:
                g[f'{c}_rollmean{W}'] = g[c].rolling(W, min_periods=max(3, W//2)).mean()

        frames.append(g)

    out = pd.concat(frames, axis=0).sort_values(keys + ['Date'])
    return out

feature_cols_base = ['CTR','CPM','CPA','Engagement_Rate','Spend_per_Day']
daily_feat = add_lags_rolls(daily_data, target='ROI_values', feat_cols=feature_cols_base, lags=(1,7), rolls=(7,))
# daily_feat.head()
```

```python
def add_cyclic_features(df):
    df['month_sin'] = np.sin(2 * np.pi * (df['month'] - 1) / 12)
    df['month_cos'] = np.cos(2 * np.pi * (df['month'] - 1) / 12)

    df['dow_sin'] = np.sin(2 * np.pi * df['dow'] / 7)
    df['dow_cos'] = np.cos(2 * np.pi * df['dow'] / 7)

    return df

daily_feat = add_cyclic_features(daily_feat)
```

```python
model_df = daily_feat.dropna().reset_index(drop=True)

display(model_df)
```

```python
clip_cols = ['ROI_values','CTR','CPM','CPA','Engagement_Rate','Spend_per_Day']
for c in clip_cols:
    q_low, q_hi = model_df[c].quantile([0.01, 0.99])
    model_df[c] = model_df[c].clip(q_low, q_hi)
```

```python
TEST_N_FUTURE_DAYS = 90

train_list, test_list = [], []
for k, g in model_df.groupby(parent_key):
    print(k)
    cutoff = g['Date'].max() - pd.Timedelta(days=TEST_N_FUTURE_DAYS)
    train_list.append(g[g['Date'] <= cutoff])
    test_list.append(g[g['Date'] > cutoff])
train = pd.concat(train_list)
test  = pd.concat(test_list)

train.shape, test.shape
```

```python
exog_cols = [c for c in model_df.columns if any(
    [c.startswith(x) for x in ['CTR','CPM','CPA','Engagement_Rate','Spend_per_Day','ROI_values_', 'dow', 'month', 'is_weekend']]
)]

print(exog_cols)
print(len(exog_cols))
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
train_exog = scaler.fit_transform(train[exog_cols])
test_exog  = scaler.transform(test[exog_cols])
```

```python
ORDER = (1,0,1)
SEASONAL_ORDER = (1,1,1,7)

def fit_and_forecast(train_df, test_df, order=ORDER, seasonal_order=SEASONAL_ORDER):
    endog_train = train_df['ROI_values'].values
    endog_test  = test_df['ROI_values'].values

    exog_train = scaler.transform(train_df[exog_cols])
    exog_test  = scaler.transform(test_df[exog_cols])

    model = SARIMAX(endog=endog_train, exog=exog_train,
                    order=order, seasonal_order=seasonal_order,
                    enforce_stationarity=False, enforce_invertibility=False)
    res = model.fit(disp=False)
    
    preds = res.get_forecast(steps=len(test_df), exog=exog_test).predicted_mean
    return res, preds

res, preds = fit_and_forecast(train, test)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

rmse = mean_squared_error(test['ROI_values'], preds)
mape = np.mean(np.abs((test['ROI_values'] - preds) / test['ROI_values'])) * 100

print(f"Test RMSE: {rmse:.4f}")
print(f"Test MAPE: {mape:.2f}%")
```

```python
for company_name in test['Company'].unique()[2:10]:
    test_sub = test[test['Company'] == company_name].copy()

    test_pos_indices = test_sub.index.to_numpy()
    pos_in_test = np.searchsorted(test.index.values, test_pos_indices)
    
    pred_sub = preds[pos_in_test]

    plt.figure(figsize=(18,6))
    plt.plot(test_sub['Date'], test_sub['ROI_values'], label='Actual ROI_values', marker='o')
    plt.plot(test_sub['Date'], pred_sub, label='Predicted ROI_values', marker='x')
    plt.xlabel('Date')
    plt.ylabel('ROI_values')
    plt.title(f'Actual vs Predicted ROI_values on Test Set ({company_name})')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
import pandas as pd

true_values = test['ROI_values'].values
pred_values = preds

mae = mean_absolute_error(true_values, pred_values)
rmse = mean_squared_error(true_values, pred_values)
mape = np.mean(np.abs((true_values - pred_values) / true_values)) * 100
r2 = r2_score(true_values, pred_values)

validation_results = pd.DataFrame({
    'Metric': ['MAE', 'RMSE', 'MAPE (%)', 'R² Score'],
    'Value': [mae, rmse, mape, r2]
})

print(validation_results)
```

```python
from xgboost import XGBRegressor
import numpy as np
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import make_scorer

scorer = make_scorer(mean_squared_error, greater_is_better=True, squared=False)

X_train = train_exog
y_train = train['ROI_values'].values

X_test = test_exog
y_test = test['ROI_values'].values

param_grid = {
    'n_estimators': [100, 300, 500],
    'learning_rate': [0.01, 0.05, 0.1],
    'max_depth': [3, 5, 7],
    'subsample': [0.7, 0.8, 1.0],
    'colsample_bytree': [0.7, 0.8, 1.0],
    'gamma': [0, 1, 5],
    'reg_alpha': [0, 0.5, 1],
    'reg_lambda': [1, 5, 10]
}

xgb = XGBRegressor(random_state=42)

search = RandomizedSearchCV(
    estimator=xgb,
    param_distributions=param_grid,
    scoring=scorer,
    n_iter=25,
    cv=3,
    verbose=1,
    random_state=42,
    n_jobs=-1
)

search.fit(X_train, y_train)
```

```python
best_model = search.best_estimator_
print("Best Parameters:", search.best_params_)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

y_pred = best_model.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100

print(f"MAE: {mae:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"R²: {r2:.4f}")
print(f"MAPE: {mape:.2f}%")
```
